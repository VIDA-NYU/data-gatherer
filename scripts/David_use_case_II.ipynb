```python
from data_gatherer.orchestrator import Orchestrator
from data_gatherer.data_fetcher import WebScraper, APIClient
import pandas as pd
import requests
import os
```

```python
dataGatherer = Orchestrator('config.json')
```

```python
# Process a list of URLs
urls = [
    "https://www.ncbi.nlm.nih.gov/pmc/articles/PMC6141466"
]
```

```python
dataGatherer.setup_data_fetcher()
results = dataGatherer.process_urls(urls)

# return the union of all the results
combined_df = pd.DataFrame()
for url, df in results.items():
    combined_df = pd.concat([combined_df, df], ignore_index=True)
```

```python
if dataGatherer.config['data_resource_preview']:
    dataGatherer.get_data_preview(combined_df,'ipynb')
```
