```python
# get new col: description_len
description_len = datasets_df['access_instructions'].apply(lambda x: len(x) if isinstance(x, str) else 0)

# plot the distribution of description_len
plt.hist(description_len, bins=50)
plt.title('Distribution of Access Instructions Lengths')
plt.xlabel('Length of Access Instructions')
plt.ylabel('Number of Datasets')
plt.show()  
```

```python
data_locations_len = []
for x in datasets_df['subject_of_study']:
    data_locations_len.append(len(x))
```

```python
# summarize data_locations_len with median, max, min
plt.boxplot(data_locations_len, vert=False)
```

```python
# min max len
min(data_locations_len), max(data_locations_len), sorted(data_locations_len)[len(data_locations_len)//2]  
```
